```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from ydata_profiling import ProfileReport
```

```python
data = pd.read_csv("AirQuality.csv", sep=";", decimal=',') # Read the file
data
```

It has a lot of empty Columns and Rows

```python
data = data.dropna(how="all").dropna(axis=1, how='all') # clean empty columns and rows
data
```

I'm just gonna remove the `Date` and `Time` Rows to create a single one.

Maybe I'm gonna create some others (Year, Month, Day and Hour) in the future.

Or maybe I'm just gonna handle it with a date-timestamp.
(I'm creating it just for fun)

```python
data["DateTime"] = data['Date'] + ' ' + data['Time']
data["DateTime"] = pd.to_datetime(data["DateTime"], format="%d/%m/%Y %H.%M.%S")
data.drop(['Date', 'Time'], axis=1, inplace=True)

# Timestamp
data["datetimestamp"] = (data['DateTime'] - pd.Timestamp("1970-01-01 00:00:00")) // pd.Timedelta('1h')
```

Date time stamp can be inverted by

```python
print(data['DateTime'][0])
print(pd.to_datetime(data["datetimestamp"][0], unit="h"))
```

And as you can see it works so well.

## -200 its a real problem
There are too many of them, and its a problem.

Spoiler `NMHC(GT)` has ~90% of its values at -200 so, I'm gonna remove it.

```python
temp_df = data.drop(['DateTime', 'datetimestamp'], axis=1)
neg_values_count = temp_df.apply(lambda x: (x < 0).sum())
```

```python
neg_values_count
```

And yep, there are too many negative values.

And just for the record, the only column that should have negative values is T, but some of them are -200, and I don't think Italy reach that ðŸ’€.

```python
temp_df.replace(to_replace = -200, value = np.nan, inplace = True)
temp_df.info()
```
