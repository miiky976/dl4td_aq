## Actually removing outliers

### Zscores del profe Jaimico

```python
from scipy import stats

def cleanZscores (df: pd.DataFrame, mayor: int, menor: int) -> pd.DataFrame :
  df = df.drop(columns=['DateTime'], errors='ignore') # eliminar temporalmente DateTime
  outliers_count = {}
  rows_before = df.shape[0]
  for column in df.columns:
    z_scores = stats.zscore(df[column])
    outliers = (z_scores > 3) | (z_scores < -3)
    outliers_count[column] = sum(outliers)
    df = df[~outliers]
  rows_after = df.shape[0]
  rows_removed = rows_before - rows_after
  print("NÃºmero de puntos atÃ­picos por variable:", outliers_count)
  print(f"Renglones eliminados: {rows_removed}")
  return df
```

```python
Q1 = data.quantile(0.25, numeric_only=True)
Q3 = data.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1

scale = 1.4
lower_lim = Q1 - scale*IQR
upper_lim = Q3 + scale*IQR
```

```python
cols = data.columns[:5]  # Look for outliers in columns starting from CO(GT)

lower_lim = lower_lim[cols]
upper_lim = upper_lim[cols]

condition = ~((data[cols] < lower_lim) | (data[cols] > upper_lim)).any(axis=1)

# Generate a new DataFrame with outliers removed
dataS = data[condition]
```
